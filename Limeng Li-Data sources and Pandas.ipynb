````python
df_mckinsey = None
df_mckinsey = None
import json
import numpy as np
import pandas as pd
import os

file_dir = '../resource/asnlib/publicdata/opencorporates/'
list_company_number = []
list_name = []
list_country = []
list_locality = []
# for file_name in os.listdir(file_dir):
def readFile(file_name):
    file_path = os.path.join(file_dir, file_name)
    with open(file_path, 'r') as fff:
        load_dict = json.load(fff)
        company = load_dict['results']['companies']
        for i in range(len(company)):
            companyInfo = company[i]['company']
            if companyInfo == None:
                list_company_number.append(np.nan)
                list_name.append(np.nan)
                list_country.append(np.nan)
                list_locality.append(np.nan)
            else:
                if companyInfo['company_number'] == None:
                    list_company_number.append(np.nan)
                else:
                    list_company_number.append(companyInfo['company_number'])
                if companyInfo['name'] == None:
                    list_name.append(np.nan)
                else:
                    list_name.append(companyInfo['name'])
                if companyInfo['registered_address'] == None:
                    list_country.append(np.nan)
                    list_locality.append(np.nan)
                else:
                    if companyInfo['registered_address']['country'] == None:
                        list_country.append(np.nan)
                    else:
                        list_country.append(companyInfo['registered_address']['country'])
                    if companyInfo['registered_address']['locality'] == None:
                        list_locality.append(np.nan)
                    else:
                        list_locality.append(companyInfo['registered_address']['locality'])
readFile("mckinsey_company_page_1.txt")
readFile("mckinsey_company_page_2.txt")
readFile("mckinsey_company_page_3.txt")
readFile("mckinsey_company_page_4.txt")
readFile("mckinsey_company_page_5.txt")
readFile("mckinsey_company_page_6.txt")
readFile("mckinsey_company_page_7.txt")
order = ['company_number', 'name', 'country', 'locality']
df_mckinsey = pd.DataFrame({ "name": list_name, "company_number": list_company_number, "country": list_country, "locality": list_locality})
df_mckinsey = df_mckinsey[order]
df_mckinsey
````

````python
import sqlite3
def create(conn):
    sql_create = '''
    CREATE TABLE 'tbl' (
        'wikipedia' TEXT NOT NULL,
        'topic' TEXT NOT NULL,
        'pageviews' INTEGER NOT NULL,
        'x' INTEGER NOT NULL
    )
    '''
    conn.execute(sql_create)

def insert(conn, wikipedia, topic, pageviews, x):
    sql_insert = '''
    INSERT INTO
        tbl
    VALUES
        (?, ?, ?, ?);
    '''
    conn.execute(sql_insert, (wikipedia, topic, pageviews, x))

path = 'db.sqlite'
conn = sqlite3.connect(path)
# conn.execute('DROP TABLE tbl')
create(conn)

p = '../resource/asnlib/publicdata/pageviews'
with open(p, 'r') as fff:
    index = 0
    for line in fff.readlines():
        w, t, p, x = line.strip().split(' ')
        index += 1
        if index == 1941 or index == 86823 or index == 99805:#############################
            t = t.replace('"','')
        insert(conn, w, t, p, x)
conn.commit()
df_pageview = pd.read_sql_query("SELECT * from tbl", conn)
df_pageview
````

3.Write and execute a sql query that creates a new table in your database called 'tbl_total'. The table should aggregates values in the table 'tbl'  and show the number of pageviews per wikipedia portal. The values in the table should be ordered lexicographically according to the portal. 

````python
# example of the output
pd.read_csv(public+'hw3_ex2_q3_sample.csv',index_col=0)
````

````python
#
# YOUR CODE HERE
#
def countForWikipedia(conn):
    sql_count = '''
    CREATE TABLE tbl_total AS
    SELECT wikipedia, sum(pageviews) 
    FROM tbl
    GROUP BY wikipedia
    ORDER BY wikipedia ASC;
    '''
    conn.execute(sql_count)
    conn.commit()

# conn.execute("DROP TABLE tbl_total")
countForWikipedia(conn)
df_count = pd.read_sql_query("SELECT * from tbl_total", conn)
df_count
````

## Part 3: Web-scraping

HTML are XML documents that specify the content and design of web-pages. For example, the following HTML code

```html
<TABLE>
    <TR>
        <TD>A</TD>
        <TD>B</TD>
        <TD>C</TD>
    </TR>
    <TR>
        <TD>D</TD>
        <TD>E</TD>
        <TD>F</TD>
    </TR>
</TABLE>
```

represents of the following table: 
<TABLE>
    <TR>
        <TD>A</TD>
        <TD>B</TD>
        <TD>C</TD>
    </TR>
    <TR>
        <TD>D</TD>
        <TD>E</TD>
        <TD>F</TD>
    </TR>
</TABLE>


The TABLE element is used to indicate the beginning of the table, TR elements indicate beginning of rows, and TD elements indicate beginning of columns. The text that appears between the TD tags specifies cells content. 

As another example, hyperlinks are specified using the A tag. A link to the website http://www.w3schools.com whose label is "Visit W3Schools.com", can be represented by the following HTML code: 
```html
<a href="http://www.w3schools.com">Visit W3Schools.com!</a> 
```

````python
from lxml import html
from lxml import etree
import pandas as pd
import numpy as np
import re
list_cake_name = []
list_cake_origin = []
with open('../resource/asnlib/publicdata/article.html', 'r') as file:
    Info = file.read()
    tree = html.fromstring(Info)
    cakeTable = tree.xpath('.//table')[0][0]
    indexRow = 0
    indexCol = 0
    replace_reg = re.compile(r'\[\d+\]')####################
    for row in cakeTable:
        if indexRow == 0:
            indexRow += 1
            continue
        indexCol = 0
        for col in row:
            if indexCol == 0:
                content = col.text_content().rstrip()
                if content == "":
                    list_cake_name.append(np.nan)
                else:
                    content = replace_reg.sub('', content)
                    list_cake_name.append(content.strip())
            elif indexCol == 2:
                content = col.text_content().rstrip()
                if content == "":
                    list_cake_origin.append(np.nan)
                else:
                    content = replace_reg.sub('', content)
                    list_cake_origin.append(content.strip())
            indexCol += 1
order_cake = ['Name', 'Origin']
df_cakes = pd.DataFrame({ "Name": list_cake_name, "Origin": list_cake_origin})
df_cakes = df_cakes[order_cake]
df_cakes
````

In this assignment, you will analyze 5 files (included in the public folder `/online learning/`) that contain information about times at which students watched online lectures. Each file name represent the topic of the lecture, and the columns within each file contain the following information: 
* Timestamp - The time at which the user started watching the lecture (e.g., 11/9/2017 8:26:44 PM)
* Start Position - The position relative to the beginning of the lecture in seconds (e.g., 15.25 implies that the student started watching 15.25 seconds after the beginning of the lecture)
* Minutes Viewed - The number of minutes viewed since 'Timestamp'
* UserID - a unique user identifier

Note that while watching movies, users can move backward and forward in time, in which case multiple records will be added to the data files. We will refer to each time interval/record as a 'stretch'.

2.Create a dataframe called "df_all" that consolidates the information from all files. The dataframe should include an additional column called "filename" which contains the source of each stretch.

````python
# example of the output
pd.read_csv(public+'hw3_ex4_q2_sample.csv',index_col=0)
````

````python
#
# YOUR CODE HERE
#
import pandas as pd
list_timeStamp = []
list_startPosition = []
list_minutesViewed = []
list_userID = []
list_filename = []
all = []
for i in range(len(files)):
    list_timeStamp_file = []
    list_startPosition_file = []
    list_minutesViewed_file = []
    list_userID_file = []
    list_filename_file = []
    filepath = '../resource/asnlib/publicdata/onlinelearning/' + files[i]
    index = 0
    with open(filepath, 'r') as f:
        for line in f.readlines():
            if index == 0:
                index += 1
                continue
            line = line.strip()
            t, s, m, u = (line.split(','))
            list_timeStamp.append(t)
            list_startPosition.append(round(float(s), 18))
            list_minutesViewed.append(float(m))
            list_userID.append(u)
            list_filename.append(files[i])
            list_timeStamp_file.append(t)
            list_startPosition_file.append(round(float(s), 18))
            list_minutesViewed_file.append(float(m))
            list_userID_file.append(int(u))
            list_filename_file.append(files[i])
    df_all_file = pd.DataFrame({"Timestamp": list_timeStamp_file, "Start Position": list_startPosition_file, "Minutes Viewed": list_minutesViewed_file, "UserID": list_userID_file, "filename": list_filename_file})
    df_all_file_order = ["Timestamp", "Start Position", "Minutes Viewed", "UserID", "filename"]
    df_all_file = df_all_file[df_all_file_order]
    all.append(df_all_file)
df_all = pd.concat(all)
df_all
````
